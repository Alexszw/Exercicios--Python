```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score

base = pd.read_csv('teste_wine .csv')

X = base[base.columns[:]]

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, n_init=8)
    agg = AgglomerativeClustering(n_clusters=k)
    kmeans_groups = kmeans.fit_predict(X)
    agg_groups = agg.fit_predict(X)
    
    print('k: %2.d kmeans: %.6f agglomerative: %.6f' % (
        k, silhouette_score(X, kmeans_groups), silhouette_score(X, agg_groups)
    ))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd ; import numpy as np; from sklearn import manifold
from matplotlib import colors, cm, ticker, pyplot as plt


base = pd.read_csv('teste_wine .csv')
X = base[base.columns[:]]
n_groups = 7


kmeans = KMeans(n_clusters=n_groups, n_init=20)
agg = AgglomerativeClustering(n_clusters=n_groups)
kmeans_groups = kmeans.fit_predict(X)
agg_groups = kmeans.fit_predict(X)


color_labels = ['#0099FF', '#009900', '#FF3300' ,'#FFD444','#FF9923','#FF8514','#FF6791' ]

fig, axes = plt.subplots(1, 2)

for ax, groups, name in zip(axes, [kmeans_groups, agg_groups], ['kmeans', 'agglomerative']):
    # projetando o k-means
    tsne = manifold.TSNE(n_components=2, init='random', random_state=0)
    transf = tsne.fit_transform(X)

    for i in np.unique(groups):  # para cada um dos grupos presentes 
        where = np.flatnonzero(groups == i)  # seleciona instâncias que estão no grupo i
        # projeta essas instâncias
        ax.scatter(transf[where, 0], transf[where, 1], c=color_labels[i], s=90, linewidth=0.)  
        ax.set_title(name)
        ax.xaxis.set_major_formatter(ticker.NullFormatter())
        ax.yaxis.set_major_formatter(ticker.NullFormatter())
    plt.axis('tight')
plt.show()
```
